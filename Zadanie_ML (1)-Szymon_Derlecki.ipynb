```python
import pandas as pd
```

```python
link = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(link)
```

```python
titanic.head()
```

```python
titanic = titanic.drop(['Name', 'Ticket', 'Cabin', 'Ticket', 'Embarked'], axis = 1)
```

```python
titanic.dtypes
```

```python
titanic['Sex'].replace(['male', 'female'], [0, 1], inplace = True)
```

```python
titanic.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
titanic[["PassengerId", "Pclass", "Age", 'SibSp', 'Fare', 'Parch']] = scaler.fit_transform(titanic[["PassengerId", "Pclass", "Age", 'SibSp', 'Fare', 'Parch']])
```

```python
titanic.head(10)
```

```python
titanic.isnull().sum()
```

```python
titanic.fillna(titanic.mean(numeric_only = True).round(1), inplace = True)
```

#**k-NN**

```python
X = titanic.drop('Survived', axis=1)
Y = titanic.Survived
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix
```

```python
kNN=KNeighborsClassifier()
kNN.fit(X_train,Y_train)
```

```python
Y_pred=kNN.predict(X_test)
```

```python
knn_accy = round(accuracy_score(Y_test, Y_pred), 3)
print (knn_accy)
print(precision_score(Y_test,Y_pred))
print(recall_score(Y_test,Y_pred))
```

```python
conf_mtx = confusion_matrix(Y_test, Y_pred)
```

```python
conf_df = pd.DataFrame(conf_mtx)
plt.figure(figsize = (5,5))
sns.heatmap(conf_df, 
           annot=True, annot_kws={"size": 25},
           fmt="d",        
           xticklabels=False, 
           yticklabels=False,
           cmap="viridis", 
           cbar=False)
plt.show()
```

#**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
reg = LogisticRegression()
reg.fit(X_train, Y_train)
```

```python
pred = reg.predict(X_test)
print(classification_report(Y_test, pred))
```

```python
conf_mtx = confusion_matrix(Y_test, pred)

conf_df = pd.DataFrame(conf_mtx)
plt.figure(figsize = (5,5))
sns.heatmap(conf_df, 
           annot=True, annot_kws={"size": 25},
           fmt="d",        
           xticklabels=False, 
           yticklabels=False,
           cmap="viridis", 
           cbar=False)
plt.show()
```

Znalezienie hiperparametrw tak ayb byo atwiej pozyska istotno cehc poniej - >

```python
logistic_param = LogisticRegression(penalty='l2', C = 10)
logistic_param.fit(X_train, Y_train)
```

#**Feature Importance**

```python
from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
```

```python
pred=X_train
sel=RFE(logistic_param, n_features_to_select=1,step=1)
sel=sel.fit(pred, Y_train)
ord=sel.ranking_

feature_ranks=[]
for ord in feature_ranks:
  feature_ranks.append(f"{ord}.{titanic.columns[ord]}")

feature_ranks
```

```python
sel = RFECV(logistic_param, step=1, min_features_to_select=1, 
                 cv=10, scoring="f1")
sel.fit(X, Y)
```

```python
chosen_features_df = pd.DataFrame({
    "chosen": sel.support_,
    "ranking": sel.ranking_
}, index=X.columns)
chosen_features_df
```
